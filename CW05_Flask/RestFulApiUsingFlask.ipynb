### Import Flask as app

```python
# demo code 
from flask import Flask
import json

app = Flask(__name__)
```

### Connect to SQL Database
> Reference: https://docs.python.org/3/library/sqlite3.html

```python
import sqlite3

dbPath = "./db/FestoMES.db"
connection = sqlite3.connect(dbPath)
connection.close()
```

### HW5 - Flask
- Step 1 ： Check the information of tblBufferPos
- Check Application: Postman
- URL -> http://127.0.0.1:5000/HW5-1
- Method -> GET
- return in Postman Console

```python
from flask import request
# Get information of all resources(machines)
@app.route("/HW5-1", methods=['GET'])
def check_resource():
    try:
        # Initialize connection with database
        conn = sqlite3.connect(dbPath)
        cur = conn.cursor()

        # Use SQL query to retrieve all resources' information
        cur.execute("SELECT * from tblBufferPos")
        
        # Get every resources' information and append on a list variable, resource_info_list
        resouceList = []
        for resourceNmae in cur:
            resouceList.append(resourceNmae)
        
        # Set headers and turn the list variable to json format
        response = app.response_class(
            response=json.dumps(resouceList),
            status=200,
            mimetype='application/json'
        )
        # Disconnect with database
        conn.close()
        return response

    except sqlite3.Error as e:
        print(e)

        # Set headers and turn a list variable to json format
        response = app.response_class(
            response=json.dumps([{'result':str(e)}]),
            status=200,
            mimetype='application/json'
        )

        return response
```

### HW5 - Flask
- Step 2 ： Change the PNo of all the information of tblBufferPos to 210, and change the TimeStamp to the current time.
- Check Application: Postman
- URL -> http://127.0.0.1:5000/HW5-2
- Method -> PUT
- JSON --> {}
- result in './db/FestoMES.db'

```python
@app.route("/HW5-2", methods=["PUT"])
def change_resource():
    resource_info_dict = request.get_json()
    try:
        # Initialize connection with database
        conn = sqlite3.connect(dbPath)
        cur = conn.cursor()

        # Use SQL query to update some part of a certain resource's information
        cur.execute("UPDATE tblBufferPos SET PNo='210', TimeStamp=CURRENT_TIMESTAMP")
        conn.commit()
        # Disconnect with database
        conn.close()
        
        # Give a result of this request
        resource_info_dict["result"] = "Successful!"

        # Set headers and turn a list variable to json format
        response = app.response_class(
            response=json.dumps(resource_info_dict),
            status=200,
            mimetype='application/json'
        )

        return response

    except sqlite3.Error as e:
        print(e)
        
        # Give a result of this request
        resource_info_dict["result"] = str(e)

        # Set headers and turn a list variable to json format
        response = app.response_class(
            response=json.dumps(resource_info_dict),
            status=200,
            mimetype='application/json'
        )

        return response
```

### HW5 - Flask
- Step 3 ： You can update the contents of the PNo of the specified Pos number, for example, change the PNo of Pos 12 to 211.
- Check Application: Postman
- URL -> http://127.0.0.1:5000/HW5-3
- Method -> POST
- JSON --> 
{
"BufPos":"12",
"PNo":"211"
}
- result in './db/FestoMES.db'

```python
@app.route("/HW5-3", methods=["POST"])
def update_s_resource():
    resource_info_dict = request.get_json()
    try:
        print(resource_info_dict)
        # Initialize connection with database
        conn = sqlite3.connect(dbPath)
        cur = conn.cursor()

        # Use SQL query to update some part of a certain resource's information
        query = f"UPDATE tblBufferPos SET PNo = '{resource_info_dict['PNo']}' WHERE BufPos = '{resource_info_dict['BufPos']}'"
        print(query)
        cur.execute(query)
        conn.commit()
        # Disconnect with database
        conn.close()
        
        # Give a result of this request
        resource_info_dict["result"] = "Successful!"

        # Set headers and turn a list variable to json format
        response = app.response_class(
            response=json.dumps(resource_info_dict),
            status=200,
            mimetype='application/json'
        )

        return response

    except sqlite3.Error as e:
        print(e)
        
        # Give a result of this request
        resource_info_dict["result"] = str(e)

        # Set headers and turn a list variable to json format
        response = app.response_class(
            response=json.dumps(resource_info_dict),
            status=200,
            mimetype='application/json'
        )

        return response
```
